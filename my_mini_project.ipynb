```python
import numpy as np  # importing numpy
import pandas as pd  # importing pandas
import matplotlib.pyplot as plt  # importing MatPlotLib

# the original signal graph
x = np.arange(0, 1, 0.01)
y = np.sin(2 * np.pi * 5 * x)
plt.plot(x, y, color='aqua', label='Original Signal')  # Plot original signal

# adding a linear line at y=0
plt.axhline(y=0, color='gray', linestyle='--')

# finding values of min and max points in original graph
max_value_indices = np.where(y == np.max(y))[0]
min_value_indices = np.where(y == np.min(y))[0]

new_values = []

# Add max and min indices to new_values
new_values.extend(max_value_indices)
new_values.extend(min_value_indices)

# downsampling the data while keeping the values of the min and max points from the original graph
for i in range(0, len(y), 5):
    if i not in new_values:
        new_values.append(i)

# extracting downsampled values
new_values = sorted(list(new_values))
new_y = y[new_values]
new_x = x[new_values]

# creating new downsampled graph while highlighting the downsamples poins taken
plt.scatter(new_x, new_y, color='deeppink', label='Downsampled Points')
plt.plot(new_x, new_y, color='hotpink', linestyle='--', label='Downsampled Signal')  # Line plot for downsampled data
# adding labels and title
plt.xlabel("time in mili-seconds")
plt.ylabel("strength of signal")
plt.title("Original Signal and Downsampled signal combined graph")
# shiowing the final plot
plt.legend()
plt.show()
```

```python
# uploading the csv file to python
dataset = pd.read_csv('laptop_price - dataset.csv')

# extracting the 'Company' and 'Price (Euro)' columns
company = dataset['Company']
prices = dataset['Price (Euro)']

# creating the scatter plot of laptop prices
plt.figure(figsize=(12, 8))
plt.scatter(company, prices, color='hotpink', edgecolors='deeppink', alpha=0.6)

# Adding titles and labels
plt.title('Laptop Prices by Company', fontsize=16)
plt.xlabel('Company', fontsize=12)
plt.ylabel('Price (Euro)', fontsize=12)

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Displaying the plot
plt.tight_layout()
plt.show()

# Calculate the average laptop price for each company
average_price_per_company = dataset.groupby('Company')['Price (Euro)'].mean().reset_index()

# Finding the company with the most expensive average laptop price
most_expensive_company = average_price_per_company.loc[average_price_per_company['Price (Euro)'].idxmax()]

# Display the average prices chart while preventing the index from being shown (it bothered me so i removed it)
print("Average Laptop Prices per Company:", average_price_per_company.to_string(index=False))  
# Display the company with the most expensive average laptop price, printing in two rows to show results in a neatly manner
print("\nCompany with the most expensive laptop average price:")
print(f"{most_expensive_company['Company']}: €{most_expensive_company['Price (Euro)']:.2f}")


# creating the plots for the price distribution of laptops for every different operation system.
# finding the different types of operation systems
different_opsys_types = dataset['OpSys'].unique()

# filtering the data and making sure that each operation system type has its own histogram plot.
for os in different_opsys_types:
    os_data = dataset[dataset['OpSys'] == os]

    # Creating the plots
    plt.figure(figsize=(10, 6))
    plt.hist(os_data['Price (Euro)'], bins=20, color='hotpink', edgecolor='deeppink', alpha=0.7)
    # adding the plots titles
    plt.title(f'The Distribution of Laptop Prices for {os}', fontsize=16)
    plt.xlabel('The Laptop Price in Euros', fontsize=12)
    plt.ylabel('The Amount Of Times The Price Appeared in Data', fontsize=12)

    # Displaying the plots
    plt.tight_layout()
    plt.show()


# Creating a scatter plot that shows the correlation between the RAM column and the Computer Price column
plt.figure(figsize=(12, 8))
plt.scatter(dataset['RAM (GB)'], dataset['Price (Euro)'], color='hotpink', edgecolors='deeppink', alpha=0.7)

# Adding a sloped line to help visualize the correlation
z = np.polyfit(dataset['RAM (GB)'], dataset['Price (Euro)'], 1)  # Linear fit
p = np.poly1d(z)
plt.plot(dataset['RAM (GB)'], p(dataset['RAM (GB)']), color='darkviolet', linestyle='--', linewidth=2, label='correlation sloped Line')

# Adding titles and labels
plt.title('Correlation between RAM and Computer Price', fontsize=16)
plt.xlabel('RAM (GB)', fontsize=12)
plt.ylabel('Computer Price (Euro)', fontsize=12)

# Displaying the plot
plt.tight_layout()
plt.show()

#the plot shows a clear correlation between the RAM and the laptop's price - the more RAM a laptop has, the more expensive it is

# Creating a new column for storage type
def extract_storage_type(memory):
    if pd.isnull(memory):
        return None
    for storage_type in ['SSD', 'HDD', 'Flash Storage', 'Hybrid']:
        if storage_type in memory:
            return storage_type
    return 'Unknown'

#creating the new column using the data that was extracted from the memory column
dataset['Storage type'] = dataset['Memory'].apply(extract_storage_type)
```
